## Dev Results visualization

```python
import matplotlib.pyplot as plt
```

```python
# Extracted data from the results
results = {
    "MLP baseline": {"MAE": 0.3717430830001831, "MSE": 0.2656225860118866, "RMSE": 0.5153858661651611, "R2": 0.5697605609893799},
    "LSTM": {"MAE": 0.35656818747520447, "MSE": 0.2506922781467438, "RMSE": 0.5006917715072632, "R2": 0.5939437747001648},
    "GRU": {"MAE": 0.36702677607536316, "MSE": 0.24354344606399536, "RMSE": 0.493501216173172, "R2": 0.6055229902267456},
    "xLSTM": {"MAE": 0.529369592666626, "MSE": 0.4781911373138428, "RMSE": 0.6915136575698853, "R2": 0.22545486688613892},
    "Mamba": {"MAE": 0.3639853298664093, "MSE": 0.22987189888954163, "RMSE": 0.4794495701789856, "R2": 0.6276673674583435},
    "Transformer": {"MAE": 0.36076900362968445, "MSE": 0.23689764738082886, "RMSE": 0.4867213368415833, "R2": 0.6162874698638916}
}

# Prepare data for plotting
models = list(results.keys())
rmse_values = [results[model]["RMSE"] for model in models]
r2_values = [results[model]["R2"] for model in models]
```

```python
# Plot RMSE
plt.figure(figsize=(10, 5))
plt.bar(models, rmse_values)
plt.title('RMSE Comparison')
plt.ylabel('RMSE')
plt.xlabel('Models')
plt.xticks(rotation=45)

plt.ylim(min(rmse_values) * 0.9, max(rmse_values) * 1.05)

plt.show()
```

```python
# Plot R2
plt.figure(figsize=(10, 5))
plt.bar(models, r2_values)
plt.title('R2 Comparison')
plt.ylabel('R2')
plt.xlabel('Models')
plt.xticks(rotation=45)

plt.ylim(min(r2_values) * 0.8, max(r2_values) * 1.05)

# Eraslan et al result
plt.axhline(y=0.58, color='r', linestyle='--')

plt.show()
```
